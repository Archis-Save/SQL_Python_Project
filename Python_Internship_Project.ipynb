#Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#Orders per Month

```python
orders = pd.read_csv("orders_per_month_2018.csv")

plt.figure()
plt.plot(orders['month'], orders['total_orders'])
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.title("Orders per Month in 2018")
plt.xticks(rotation=45)
plt.show()
```

#Cumulative Monthly Sales

```python
df = pd.read_csv("cumulative_sales.csv")

plt.figure()
plt.plot(df['month'], df['cumulative_sales'])
plt.xlabel("Month")
plt.ylabel("Cumulative Sales")
plt.title("Cumulative Monthly Sales")
plt.show()
```

#Customer Retention Rate

```python
df = pd.read_csv("customer_retention.csv")

rate = df['retention_rate_percentage'][0]

plt.figure()
plt.bar(["Retention Rate"], [rate])
plt.ylabel("Percentage")
plt.title("Customer Retention Rate (6 Months)")
plt.ylim(0, 5)
plt.show()
```

#Top Customers

```python
df = pd.read_csv("top_customers.csv")

plt.figure()
plt.bar(df['customer_unique_id'], df['total_spent'])
plt.xlabel("Customer ID")
plt.ylabel("Total Spending")
plt.title("Top Customers by Total Spend")
plt.xticks(rotation=90)
plt.show()
```

#Top Seller

```python
sellers = pd.read_csv("top_sellers.csv")

plt.figure()
plt.bar(sellers['product_id'], sellers['total_revenue'])
plt.xlabel("Seller ID")
plt.ylabel("Total Revenue")
plt.title("Top Sellers by Revenue")
plt.xticks(rotation=90)
plt.show()
```
